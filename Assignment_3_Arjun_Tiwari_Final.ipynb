# Linear Model (Logistic Regression)

## OLS Regression
We can use Ordinary Least Square Regression to find dependency between the target variable(Dependent) and the independent variables. In a linear regression model with fixed level-one effects and a set of explanatory variables, the ordinary least squares (OLS) method selects the unknown parameters by minimizing the average of the squares of the variations between the observed dependent variable's values in the input dataset and the result of the (linear) feature of the independent variable. (Wikipedia Definition)


```python
# Every independant feature(Predictors will be matched aginst dependent feature to find there dependency )
import statsmodels.api as sm
model = sm.OLS(dataSet['Diabetes'], dataSet[['HighBP', 'Stroke', 'DiffWalk', 'PhysHlth', 'MentHlth',
       'GenHlth', 'HvyAlcoholConsump','Veggies', 'Fruits', 'PhysActivity','HeartDiseaseorAttack','Smoker','BMI','CholCheck','HighChol','Sex','Age']]).fit()
model.summary()
```

## Interpretting Shap values

```python
#In regression analysis, logistic regression (or logit regression) is estimating the parameters of a logistic model (the coefficients in the linear combination)
#wikipedia definition
linear_regression_model = models.get('logisticRegression').fit(x_train,y_train)
linear_regression_explainer = shap.LinearExplainer(linear_regression_model, x_train, feature_perturbation="interventional")
linear_regression_shap_values = linear_regression_explainer.shap_values(x_test)
shap.summary_plot(linear_regression_shap_values, x_test, feature_names=list(x_train.columns))
```

## RFE

```python
#Recursive feature elimination (RFE) is a feature selection method that fits a model and removes the weakest feature (or features) until the specified number of features is reached.
# Google definiton
logisticRegression = models.get('logisticRegression')
lr_RFE = RFE(logisticRegression, n_features_to_select=8)
```

```python
_ = lr_RFE.fit(x_train, y_train)
print(x_train.columns[lr_RFE.support_])
```

## LIME

```python
#LIME, the acronym for local interpretable model-agnostic explanations, is a technique that approximates any black box machine learning model with a local, interpretable model to explain each individual prediction.
explainer_lr_lime = lime.lime_tabular.LimeTabularExplainer(x_train.values, feature_names=x_train.columns, class_names=['0', '1'], discretize_continuous=False)

observation_lr = x_train.iloc[0]
explaination_lr = explainer_lr_lime.explain_instance(observation_lr.values, lr_RFE.predict_proba, num_features=10)

# Print the LIME values
print(explaination_lr.as_list())
```

```python
fig_lime_lr = explaination_lr.as_pyplot_figure()
plt.show()
```
