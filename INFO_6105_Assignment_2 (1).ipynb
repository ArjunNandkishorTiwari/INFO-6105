The H2O Python API includes a class called H2OGeneralizedLinearEstimator, which is used to build generalized linear models (GLMs) in H2O. 
GLMs are a type of linear model that can handle both continuous and categorical data and can be used to solve regression or classification problems.
Hyperparameters used:-
1. Family (Binomial):- Family parameter is used to specify the type of generalized linear model that needs to be fitted.
2. n folds :- This specifies the number of folds to be used for cross validation
3. Seed :- This parameter is used to control the randomness on the training process
4. compute_p_values :- This model computes whether to compute p values or not for model_coefficiants
5. remove_collinear_columns :- This parameter will remove columns if collinearity is detected
6. score_each_iteration :- This parameter is used to score each iteration
7. lambda :- This parameter is used to specify the type and strength of regularization

If P-Value is greater then 0.05 -> Variables are important statistically Else -> Variables are not important statistically

# Does regularization help?
In H2O AutoML, regularization is the process of introducing a penalty term to the loss function in order to avoid overfitting and enhance the model's generalization capabilities. When a machine learning model is trained to fit the training data too closely, overfitting takes place. This might make the model perform poorly on new, unforeseen data.

L1/ Ridge (alpha = 0) and L2/Lasso (alpha = 1) regularization, which impose penalties proportional to the absolute and squared values of the model parameters, respectively, are two regularization approaches used by H2O AutoML to avoid overfitting. These penalties drive the model to discover patterns in the data that are easier to understand and simpler, which can aid in the model's ability to generalize to new data.

```python
accuracy=[]
loss=[]
alpha_values = [0.1,0.5,0.9]
for i in alpha_values:

  glm_new_model = H2OGeneralizedLinearEstimator(family= "binomial",seed=2022,alpha=i)
  glm_new_model.train(train_columns, "Diabetes", training_frame= h2o_train)
  accuracy.append(glm_new_model.model_performance(h2o_test)['AUC'])
  loss.append(glm_new_model.model_performance(h2o_test)['logloss'])
  print("Alpha = ",i)
  glm_new_model.show()
```

```python
accuracy_1=[]
loss_1=[]
alpha_values = [0.1,0.5,0.9]
lambda_values = [0.001, 0.01, 0.1]
for i in alpha_values:
  for j in lambda_values:

    glm_new_model = H2OGeneralizedLinearEstimator(family= "binomial",seed=2022,alpha=i, lambda_=j)
    glm_new_model.train(train_columns, "Diabetes", training_frame= h2o_train)
    accuracy_1.append(glm_new_model.model_performance(h2o_test)['AUC'])
    loss_1.append(glm_new_model.model_performance(h2o_test)['logloss'])
    print("Alpha = ",i," Lambda = ",j)
    glm_new_model.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 10))
axs[0].plot(range(len(accuracy_1)), accuracy_1, '-o')
axs[0].set_xticks(range(len(accuracy_1)))
axs[0].set_xticklabels(['lambda={}, alpha={}'.format(lambda_values[i//3], alpha_values[i%3]) for i in range(len(accuracy_1))], rotation=45, ha='right')
axs[0].set_xlabel('Regularization Combination')
axs[0].set_ylabel('Area Under Curve')
axs[1].plot(range(len(loss_1)), loss_1, '-o')
axs[1].set_xticks(range(len(loss_1)))
axs[1].set_xticklabels(['lambda={}, alpha={}'.format(lambda_values[i//3], alpha_values[i%3]) for i in range(len(loss_1))], rotation=45, ha='right')
axs[1].set_xlabel('Model')
axs[1].set_ylabel('Log Loss')
plt.tight_layout()
plt.show()
```
