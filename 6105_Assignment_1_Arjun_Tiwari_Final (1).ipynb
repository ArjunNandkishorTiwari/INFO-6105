## OLS Regression

We can use Ordinary Least Square Regression to find dependency between the target variable(Dependent) and the independent variables. 
In a linear regression model with fixed level-one effects and a set of explanatory variables, the ordinary least squares (OLS) method selects the unknown parameters by minimizing the average of the squares of the variations between the observed dependent variable's values in the input dataset and the result of the (linear) feature of the independent variable. (Wikipedia Definition)

```python
# Every independant feature(Predictors will be matched aginst dependent feature to find there dependency )
import statsmodels.api as sm
model = sm.OLS(dataSet['Diabetes'], dataSet[['HighBP', 'Stroke', 'DiffWalk', 'PhysHlth', 'MentHlth',
       'GenHlth', 'HvyAlcoholConsump','Veggies', 'Fruits', 'PhysActivity','HeartDiseaseorAttack','Smoker','BMI','CholCheck','HighChol','Sex','Age']]).fit()
model.summary()
```

## Chi-Squared Test
We can use Chi Squared Test to find dependency relation between categorical variables.

If P-Value is greater then 0.05 -> Variables are Independent
Else -> Variables are Dependent

```python
from scipy.stats import chi2_contingency
import pandas as pd

categoricalFeatures = ['HighBP', 'Stroke', 'DiffWalk', 'PhysHlth', 'MentHlth','GenHlth', 'HvyAlcoholConsump','Veggies', 'Fruits', 'PhysActivity','HeartDiseaseorAttack','Smoker','CholCheck','HighChol','Sex','Age']

for i in range(0, len(categoricalFeatures)):
  for j in range(i+1, len(categoricalFeatures)):
    contigencyTable = pd.crosstab(dataSet[categoricalFeatures[i]], dataSet[categoricalFeatures[j]])

    stat, p_value, dof, expected = chi2_contingency(contigencyTable)


    
    print(categoricalFeatures[i], " | ",categoricalFeatures[j], " | P Value = ", p_value )


```

# Models

```python
# Will use standard  Classification models will be used as this is a classification problem
# Results from this set of prediction will be used as a benchmark for other predicitons after processing on the dataset
# These results are calculated when the dataset is in its purest form without any alteration
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import f1_score, confusion_matrix, ConfusionMatrixDisplay

# Using a dictionary to store the models 
# Dictionary/ Map has a O(1) look up
# As these model classes will be stored only once and can be used over and over again without the need to be imported/invoked/ defined again
models = {
    'decisionTree' : DecisionTreeClassifier(random_state=42),
    'randomForest' : RandomForestClassifier(random_state=42),
    'logisticRegression' : LogisticRegression(max_iter = 50000,random_state=42 ),
    'naiveGaussian' : GaussianNB()
}
```

## Removing Outliers

### Z - Score method
#### It is useful to standardized the values (raw scores) of a normal distribution by converting them into z-scores because.A z-score tells you where the score lies on a normal distribution curve


#### BMI


```python
dataSet_outlier = pd.read_csv("https://raw.githubusercontent.com/ArjunNandkishorTiwari/INFO-6105/main/diabetes_data.csv")
mean_bmi = dataSet_outlier['BMI'].mean()
std_bmi = dataSet_outlier['BMI'].std()

z_score = (dataSet_outlier['BMI'] - mean_bmi) / std_bmi
outliers = dataSet_outlier[abs(z_score) > 1]

z_scored_dataSet = dataSet_outlier[~dataSet_outlier.isin(outliers)].dropna()
z_scored_dataSet.head()
```
